```python
dataset.dropna(inplace=True)
```

### Duplicate Values

```python
dataset.duplicated()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.info()
```

## Data Analysis (EDA)

```python
df = dataset[dataset['price']<2000]
sns.boxplot(data=df, x='price')
```

### Price Distribution

```python
sns.histplot(data=df, x='price', bins=100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.xlabel("Price")
plt.show()
```

```python
df.columns
```

### Availability Distribution

```python
sns.histplot(data=df, x='availability_365')
plt.title("Availability Distribution")
plt.ylabel("Frequency")
plt.xlabel("Availability")
plt.show()
```

### Neighbourhood Prices

```python
df1 = df.groupby(['neighbourhood_group'])['price'].mean()
```

```python
df1
```

```python
ax = sns.barplot(x=df1.index, y=df1.values)
ax.bar_label(ax.containers[0])
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.title('Average Price per Neighbourhood Group')
plt.show()
```

### Bivariate Analysis

```python
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.show()
```

### Price & Reviews

```python
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.title("Price & Reviews around Neighbourhood Group")
plt.show()
```

### Room Types

```python
sns.countplot(x='room_type', data=df)
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

### Listings in Neighbourhood

```python
sns.countplot(y='neighbourhood_group', data=df)
plt.title('Number of Listings by Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

### Price as per Room Type

```python
sns.barplot(data=df, x='room_type', y='price', hue='room_type')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.show()
```

### Review

```python
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
reviews_by_time = df.groupby(df['last_review'].dt.to_period('M')).size()
```

```python
reviews_by_time.plot(kind='line', color='blue')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

### Neighbourhood Average Prices

```python
df2 = df.groupby(['neighbourhood'])['price'].mean().reset_index()
```

```python
plt.figure(figsize=(6,35))
sns.barplot(data=df2, y='neighbourhood', x='price')
plt.show()
```
